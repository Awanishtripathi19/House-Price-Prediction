```python
import pandas as pd
import numpy as np
from sklearn import datasets
```

```python
boston_dataset = pd.read_csv("HousingData.csv")
```

```python
boston_dataset.head()
```

```python
boston_dataset.info()
```

```python
boston_dataset.describe()
```

```python
boston_dataset.isnull().sum()
```

```python
boston_dataset['CRIM'].fillna(boston_dataset['CRIM'].median(),inplace = True)
```

```python
boston_dataset['ZN'].fillna(boston_dataset['ZN'].median(),inplace = True)
boston_dataset['INDUS'].fillna(boston_dataset['INDUS'].median(),inplace = True)
boston_dataset['CHAS'].fillna(boston_dataset['CHAS'].median(),inplace = True)
boston_dataset['AGE'].fillna(boston_dataset['AGE'].mean(),inplace = True)
boston_dataset['LSTAT'].fillna(boston_dataset['LSTAT'].median(),inplace = True)
```

```python
boston_dataset.isnull().sum()
```

```python
boston_dataset.describe()
```

```python
X = boston_dataset.drop('MEDV',axis = 1)
Y = boston_dataset['MEDV']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.2, random_state = 2)
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x_train_scaled,y_train)
```

```python
y_pred = model.predict(x_test_scaled)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_pred, y_test)
```

```python
print(mse)
```
